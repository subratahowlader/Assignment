# Task 1  
Temperature Check  
Problem:  
Write a program that takes the current temperature as input and prints:  
● "It's too cold!" if below 10°C  
● "It's cool outside" if between 10°C and 25°C  
● "It's hot outside!" if above 25°C  

```python
temp = float(input("Enter the current temperature: "))
if temp < 10:
    print("It's too cold!")
elif 10 <= temp <= 25:
    print("It's cool outside")
else:
    print("It's hot outside!")
```

# Task 2  

ATM Withdrawal System  
Problem:    
Ask the user for their account balance and the amount they want to withdraw.  
● If the withdrawal amount is greater than the balance → print "Insufficient Balance"  
● If the withdrawal amount is less than or equal to balance → print "Transaction  
Successful" and show the remaining balance.

```python
balance = float(input("Enter your account balance: "))
withdrawal_amount = float(input("Enter the amount you want to withdraw: "))

if withdrawal_amount > balance:
    print("Insufficient Balance")
else:
    balance -= withdrawal_amount
    print("Transaction Successful")
    print(f"Remaining Balance: {balance} TK")
```

# Task 4  
Bus Ticket Discount  
Problem:  
Ask the user for age and print ticket price:  
● Below 5 → Free  
● 5–18 → 50% Discount  
● 60 or above → 30% Discount  
● Others → Full Price 

```python
age = float(input("Please enter your age: "))
price = 100  
if age < 5:
    discount_price = 0
elif 5 <= age <= 18:
    discount_price = price * 0.5
elif age >= 60:
    discount_price = price * 0.7
else:
    discount_price = price

print("Ticket price:", discount_price)
```

# Task 5
Restaurant Bill with Discount  
Problem:  
Take total bill amount from the user:  
● If the bill is above 1000 → apply 10% discount  
● Otherwise → no discount  
Then print the final bill amount.  

```python
bill = float(input("Please give total bill"))
if bill > 1000:
    discount = bill * 0.10
    amount = bill - discount
else:
    amount = bill

print("Final bill amount:", amount)
```

# Task 8
Leap Year Checker  
Problem:  
Take a year as input and check:  
● If it is divisible by 400 → Leap Year  
● Else if divisible by 4 but not by 100 → Leap Year  
● Otherwise → Not a Leap Year  

```python
year = int(input("Please enter a year: "))
if year % 400 == 0:
    print("Leap Year")
elif year % 4 == 0 and year % 100 != 0:
    print("Leap Year")
else:
    print("Not a Leap Year")
```

# Task 10
Electricity Bill Calculator  
Problem:  
Input number of units consumed and calculate bill:  
● Up to 100 units → 5 BDT per unit  
● 101–200 units → 7 BDT per unit  
● Above 200 units → 10 BDT per unit  
Print total payable amount.
