# Task 01
Email Validator (String Filtering + Loops)  
Real-world use: Used in signup forms or data cleaning tasks.  
Task:  
Check if a given email is valid ‚Äî must contain ‚Äú@‚Äù and end with ‚Äú.com‚Äù or ‚Äú.org‚Äù.  
Example:  
Input: hello@inceptionbd.com  
Output: Valid Email  

```python
email = input("Enter your email: ")

a = False
b = False

for i in email:
    if i == '@':
        a = True
        break

if email.endswith(".com") or email.endswith(".org"):
    b = True

if a and b:
    print("Valid Email")
else:
    print("Invalid Email")
```

# Task 02
Count Word Frequency in Customer Reviews  
üíº Real-world use: Analyzing reviews or comments in text analytics.  
Task:  
Given a review text, count how many times each word appears (ignore case).  
Example:  
Input: "This product is good and this service is excellent"  
Output: {'this': 2, 'product': 1, 'is': 2, 'good': 1, 'and': 1, 'service': 1, 'excellent': 1}  

```python
text = input("Enter your review: ")
text = text.lower()
words = text.split()
counter = {}

for word in words:
    if word in counter:
        counter[word] += 1
    else:
        counter[word] = 1

print(counter)
```

# Task 03
Password Strength Checker  
üíº Real-world use: Used in user authentication systems.  
Task:  
Check if a password is strong ‚Äî must have at least one uppercase, one lowercase, one digit, and one special character.  
Example:  
Input: "Hello@123"  
Output: Strong Password  

```python
password = input("Enter your password: ")

if (any(i.isupper() for i in password) and
    any(i.islower() for i in password) and
    any(i.isdigit() for i in password) and
    any(not i.isalnum() for i in password)):
    print("Strong Password")
else:
    print("Weak Password")
```

# Task 04
Username Generator
üíº Real-world use: Automatically create usernames for users.  
Task:  
Take user‚Äôs full name and generate a short username.  
Rules: first 3 letters of first name + last 3 letters of last name + random number (use loops only).  
Example:  
Input: "Mamun Malitha"  
Output: mamitha57

```python
import random
name = input("Enter full name: ").split()
firstname, lastname = name[0], name[-1]

a = ""
count = 0
for i in firstname:
    a += i.lower()
    count += 1
    if count == 3:
        break

b = ""
count = 0
for i in reversed(lastname):
    b = i.lower() + b
    count += 1
    if count == 3:
        break

num = random.randint(10, 99)
print(a + b + str(num))
```

# Task 05
SMS Word Counter  
üíº Real-world use: Used in messaging apps or social media post limit check.  
Task:  
Count how many characters and words are in a message, and tell if it exceeds 160 characters (SMS limit).  
Example:  
Input: "This is a demo message."  
Output: Words = 5, Characters = 23, Status = Within Limit  

```python
message = input("Enter your message: ")

char_count = 0
for i in message:
    char_count += 1

word_count = 0
word = False
for i in message:
    if i != " " and not word:
        word_count += 1
        word = True
    elif i == " ":
        word = False

status = "Within Limit" if char_count <= 160 else "Exceeded Limit"

print(f"Words = {word_count}, Characters = {char_count}, Status = {status}")
```

# Task 06
Remove Duplicates from Contact List  
üíº Real-world use: Data cleaning in CRM or user databases.  
Task:  
Given a list of names (comma separated), remove duplicates.  
Example:  
Input: "Rahim, Karim, Rahim, Sultana"  
Output: "Rahim, Karim, Sultana"

```python
names = input("Enter names (comma separated): ").split(',')
a = []

for i in names:
    j = i.strip()
    if j not in a:
        a.append(j)

print(', '.join(a))
```

# Task 07
Detect Suspicious Words in Text (Moderation)  
üíº Real-world use: Used in chat moderation or email filtering.  
Task:  
If any ‚Äúbanned‚Äù words like ['spam', 'scam', 'fake'] are found in a text, print ‚ÄúWarning‚Äù.  
Example:  
Input: "This is a scam offer"  
Output: Warning: Suspicious content detected  

```python
text = input("Enter text: ").lower()
banned = ['spam', 'scam', 'fake']
warning = False

for i in banned:
    if i in text:
        warning = True
        break

if warning:
    print("Warning: Suspicious content detected")
else:
    print("No issues found")
```

# Task 10
Word Reversal for Data Encryption  
üíº Real-world use: Simple text obfuscation or playful encoding.  
Task:  
Reverse each word in a sentence while keeping word order the same.  
Example:  
Input: "AI is fun"  
Output: "IA si nuf"

```python
sentence = input("Enter a sentence: ")

reversed_sentence = ' '.join(word[::-1] for word in sentence.split())

print(reversed_sentence)
```
